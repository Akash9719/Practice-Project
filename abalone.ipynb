```python
from scipy.stats import zscore
z = np.abs(zscore(df))
df_new = df[(z<3).all(axis = 1)]
```

```python
df= df_new
```

```python
x= df.drop('Rings',axis=1)
y= df['Rings']
```

```python
#Now lets remove the skewness
for index in x.skew().index:
    if x.skew().loc[index]>0.5:
        x[index]=np.log1p(x[index])
    if x.skew().loc[index]<-0.5:
        x[index]=np.square(x[index])
```

```python
#train the model to get the best random state:
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
def maxr2_score(regr,x,y):
    max_r_score=0
    for r_state in range(42,100):
        x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = r_state,test_size=0.30)
        regr.fit(x_train,y_train)
        pred = regr.predict(x_test)
        r2_scr=r2_score(y_test,pred)
        print("r2 score corresponds to ",r_state," and r2 score is ",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("max r2 score corresponds to ",final_r_state," and r2 score is ",max_r_score)
    return final_r_state
```

```python
#getting the r2 score using decision tree
from sklearn.tree import DecisionTreeRegressor

dtr=DecisionTreeRegressor()
r_state=maxr2_score(dtr,x,y)
```

```python
#checking the cross validation score using decision tree regressor
from sklearn.model_selection import cross_val_score

cvs = cross_val_score(dtr,x,y,cv=5)
print('cross val score',cvs.mean())
```

```python
#checking r2 score using random forest regressor
from sklearn.ensemble import RandomForestRegressor

rfr=RandomForestRegressor()
r_state=maxr2_score(rfr,x,y)
```

```python
#checking the cross validation score using random forest regressor
cvs = cross_val_score(rfr,x,y,cv=5)
print('cross val score',cvs.mean())
```

```python
#checking r2 score using kneighbors regressor
from sklearn.neighbors import KNeighborsRegressor

knn=KNeighborsRegressor()
r_state=maxr2_score(knn,x,y)
```

```python
#checking the cross validation score using random forest regressor
cvs = cross_val_score(knn,x,y,cv=5)
print('cross val score',cvs.mean())
```

```python
#checking r2 score using svc
from sklearn.svm import SVR

svr=SVR()
r_state=maxr2_score(svr,x,y)
```

```python
#checking the cross validation score using support vector regressor
cvs = cross_val_score(knn,x,y,cv=5)
print('cross val score',cvs.mean())
```

```python
from sklearn.ensemble import AdaBoostRegressor

ada=AdaBoostRegressor()
r_state=maxr2_score(ada,x,y)
```

```python
#checking the cross validation score using adaboost classifier
cvs = cross_val_score(ada,x,y,cv=5)
print('cross val score',cvs.mean())
```

random forest regressor is the best fit model for abalone

```python
#hyper parameter tunning of r2 score using grid search cv
from sklearn.model_selection import GridSearchCV

rfr=RandomForestRegressor()
parameters = {"n_estimators":[10,100,500]}
gcv = GridSearchCV(rfr, parameters, cv=5)
gcv.fit(x,y)
gcv.best_params_
```

```python
#checking r2 score using random forest regressor
from sklearn.ensemble import RandomForestRegressor

rfr=RandomForestRegressor(n_estimators=500)
r_state=maxr2_score(rfr,x,y)
```
